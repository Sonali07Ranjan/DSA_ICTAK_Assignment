### 1. Reading the iris data set

```python
import pandas as pd
iris_data=pd.read_excel('iris.xls')
iris_data
```

### 2. Display the columns in the dataset

```python
print(iris_data.columns)
```

### 3. Calculate the mean of each column

```python
df_mean=iris_data[['SL', 'SW', 'PL', 'PW']].mean()
print(df_mean)
```

### 4. Check for the null values present in the dataset

```python
#checking for null values present in the entire dataset using isnull()
print(iris_data.isnull())
#to check overall if any null value is present in the data set
print('\n\nAny null value present in dataset: ',iris_data.isnull().any().any())
```

### 5. Visualisations

####  1) Scatter plots for SW vs SL and PW vs PL for each classification of iris

```python
import matplotlib.pyplot as plt
iris_setosa=iris_data[iris_data['Classification']=='Iris-setosa']
iris_versicolor=iris_data[iris_data['Classification']=='Iris-versicolor']
iris_virginica=iris_data[iris_data['Classification']=='Iris-virginica']
plt.figure(figsize=(10,10))

plt.subplot(3,2,1)
plt.scatter(x='SL',y='SW',c='blue',data=iris_setosa)
plt.title('SW vs SL for species setosa',fontsize=15)
plt.xlabel('SL')
plt.ylabel('SW')

plt.subplot(3,2,2)
plt.scatter(x='PL',y='PW',c='blue',data=iris_setosa)
plt.title('PW vs PL for species setosa',fontsize=15)
plt.xlabel('PL')
plt.ylabel('PW')

plt.subplot(3,2,3)
plt.scatter(x='SL',y='SW',c='red',data=iris_versicolor)
plt.title('SW vs SL for species versicolor',fontsize=15)
plt.xlabel('SL')
plt.ylabel('SW')

plt.subplot(3,2,4)
plt.scatter(x='PL',y='PW',c='red',data=iris_versicolor)
plt.title('PW vs PL for species versicolor',fontsize=15)
plt.xlabel('PL')
plt.ylabel('PW')

plt.subplot(3,2,5)
plt.scatter(x='SL',y='SW',c='green',data=iris_virginica)
plt.title('SW vs SL for species virginica',fontsize=15)
plt.xlabel('SL')
plt.ylabel('SW')

plt.subplot(3,2,6)
plt.scatter(x='PL',y='PW',c='green',data=iris_virginica)
plt.title('PW vs PL for species virginica',fontsize=15)
plt.xlabel('PL')
plt.ylabel('PW')

plt.tight_layout()
```

#### 2) Violin plot for SW, SL, PL, PW with respect to each classification

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
sns.violinplot(x='Classification',y='SL',data=iris_data)

plt.subplot(2,2,2)
sns.violinplot(x='Classification',y='SW',data=iris_data)

plt.subplot(2,2,3)
sns.violinplot(x='Classification',y='PL',data=iris_data)

plt.subplot(2,2,4)
sns.violinplot(x='Classification',y='PW',data=iris_data)

plt.tight_layout()
```
