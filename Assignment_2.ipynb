### 1. Read the dataset to the python environment

```python
import pandas as pd
df1=pd.read_csv('cars_data.csv')
df1
```

### 2. Check for the null values present in the dataset

```python
df1.isnull().sum() ## gives column-wise count of occurances of null values
```

```python
df1.isnull().sum().sum() ## gives count of total occurances of null values in given dataset
```

### 3. Plot a bar graph of male vs female buyers participated in the sales

```python
import seaborn as sns
sns.catplot(x='Buyer Gender', kind='count', data=df1).set(title="Bar plot for male and female buyers")
```

### 4. Find the top 5 cars based on their sales price

```python
df2=df1.copy()
df2['Car']=df2['Make'].astype(str)+' '+df2['Model'] ## combining make and model to get the car
## adding a new column for ranking as per sales price
df2['Rank on Sales price'] = df2.sort_values(by=['Sale Price'], ascending=False) \
               .reset_index() \
               .sort_values('index') \
               .index + 1
df3=df2.copy()
df3=df3.sort_values(by=['Rank on Sales price'])
df3=df3[['Car','Sale Price','Rank on Sales price']]
df3.head() ## displays the top 5 cars based on Sales price
```

### 5. Find the least 5 cars based on their Resell price

```python
## adding a new column for reverse ranking as per resell price
df2['Reverse Rank on Resell price'] = df2.sort_values(by=['Resell Price']) \
               .reset_index() \
               .sort_values('index') \
               .index + 1
df4=df2.copy()
df4=df4.sort_values(by=['Reverse Rank on Resell price'])
df4=df4[['Car','Resell Price','Reverse Rank on Resell price']]
df4.head() ## displays the last 5 cars based on Resell price
```
